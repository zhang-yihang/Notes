```python
import requests
import re
```

```python
website="https://dytt89.com"
son_list=[]
```

```python
resp=requests.get(website,verify=False) # 去掉安全验证
resp.encoding = "gb2312"
print(resp.text)
```

```python
obj1=re.compile(r"必看热片.*?<ul>(?P<ul>.*?)</ul>",re.S)
obj2=re.compile(r"href='(?P<son_url>.*?)'",re.S)
```

```python
ret1=obj1.finditer(resp.text)
for i in ret1:
    print(i.group("ul"))
    box=i.group("ul")
    ret2=obj2.finditer(box)
    for i in ret2:
        print(i.group("son_url"))
        son_url=i.group("son_url")
        son_web=website+son_url
        son_list.append(son_web)
```

```python
for son in son_list:
    resp2=requests.get(son,verify=False) # 去掉安全验证
    resp2.encoding = "gb2312"
    print(resp2.text)
    break
```
